```python
import os

os.environ["LD_LIBRARY_PATH"] = "/home/renan/MATLAB/R2021b/runtime/glnxa64"

import numpy as np
import BTD3
import myDehankelization
import myHankelization
import seaborn as sns
import matplotlib.pyplot as plt
import matlab
import pandas as pd
from random import randint
from scipy.io import savemat
from scipy.signal import resample
import ecg_plot
```

```python
with open(f'tensor_data/sample100hzRank6/leads.npy', 'rb') as f:
    sample = np.load(f)
sample.shape
```

```python
ecg_plot.plot(sample/1000, sample_rate=100, title='')
```

```python
T = np.zeros((50, 51, 12))
my_hankelization = myHankelization.initialize()
for i in range(0, 12):
    vec = matlab.double(sample[i].tolist(), size=sample[0].shape)
    T[:, :, i] = my_hankelization.hankelization(vec)
my_hankelization.terminate()

T = matlab.double(T.tolist(), size=T.shape)
R = matlab.double([6.0], size=(1, 1))
multirank = matlab.double([40.0, 40.0, 1.0], size=(1, 3))
my_BTD = BTD3.initialize()
output = my_BTD.BTD(R, multirank, T)
my_BTD.terminate()

output_np = np.array(output)
sources = np.empty((0, 100), int)
contrib_sources = np.empty((0, 12), int)
my_dehankelization = myDehankelization.initialize()
for i in range(0, len(output_np)):
    a = np.array(output_np[i][0])
    b = np.array(output_np[i][1]).T
    contrib_sources = np.append(contrib_sources, np.array(output_np[i][2]))
    mixing_matrix = np.dot(a, b)
    m_hankel = matlab.double(mixing_matrix.tolist(), size=mixing_matrix.shape)
    source = my_dehankelization.dehankelization(m_hankel)
    sources[i, :] = np.array(source).T
my_dehankelization.terminate()
```

```python
x = np.linspace(0, 1, 100)
for i in range(0, 6):
    y = sources[i]
    plt.figure()
    sns.lineplot(x, y)
    plt.savefig(f'tensor_data/sample100hzRank6/source_{i}.png')
```
