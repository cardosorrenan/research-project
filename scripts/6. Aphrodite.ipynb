### Caracteristicas Afrodite

```python
files = os.listdir('../workdata/shura/experiment2/rank_5/')
n_files = len(files)

df = pd.DataFrame(columns=['id', 'data', 'label'])

for i, filename in enumerate(files):
    print(f"{round((i*100)/n_files, 2)}%")
    with open(f'../workdata/shura/experiment2/rank_5/{filename}', 'rb') as f:
        output = np.load(f, allow_pickle=True)
    output = np.array(output)
    
    I = np.array(output[0][0]).shape[0]
    J = np.array(output[0][1]).shape[0]
    N = I*2 - 1 if I == J else I * 2

    sources = np.empty((0, N), float)
    contrib_sources = np.empty((0, 12), float)
    my_dehankelization = myDehankelization.initialize()
    for i in range(0, len(output)):
        a = np.array(output[i][0])
        b = np.array(output[i][1]).T
        c = np.array(output[i][2]).T
        contrib_sources = np.append(contrib_sources, c, axis=0)
        mixing_matrix = np.dot(a, b)
        m_hankel = matlab.double(mixing_matrix.tolist(), size=mixing_matrix.shape)
        source = my_dehankelization.dehankelization(m_hankel)
        sources = np.append(sources, np.array(source).T, axis=0)
    my_dehankelization.terminate()
    label = base_df.loc[int(filename.split('.')[0])].diagnostic
    result = {'id': int(filename.split('.')[0]),
              'data': sources,
              'label': label}
    df = df.append(result, ignore_index=True)
df.to_pickle('../workdata/aphrodite/sources_china.pkl')
```

```python
df = pd.read_pickle('../workdata/aphrodite/sources_china.pkl')

# Sources extracted
S = 5

# Get the min size source between the samples
N = int(np.mean(list(map(lambda x: x.shape[1], df.data)))) # 96

for i, row in df.iterrows():
    sources = np.empty((S, N), float)
    for j, source in enumerate(row.data):
        source = source.reshape(1, source.shape[0])
        sources[j] = resample(source, N, axis=1)
    df.at[i, 'data'] = sources
  
Y = df[['label']]
Y = Y['label'].to_numpy()
le.fit(Y)
Y = le.transform(Y)

n_rows, _ = df.shape
```
